### **5. Cleaning the dataset**

```python
#Dropping the object data type columns as those are of no use to us.
iris1 = iris.drop(columns = ['Species','Id'])
iris1.head() # See the first 5 rows
```

<div id='6'></div>

<br>

### **6. Importing and running K means clustering.**

```python
#Importing K means from sklearn.cluster
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)

#Fitting and predicting the present clusters
kmeans.fit(iris1)
pred=kmeans.predict(iris1)
```

```python
#Array of all the elements in the two clusters(0 and 1)
pred, len(pred)
```

```python
#Number of zero clusters and one clusters.
pd.Series(pred).value_counts()
```

```python
#Inertia of this prediction
kmeans.inertia_
```

```python
#Score of this prediction
kmeans.score(iris1)
```

<div id='7'></div>

<br>

### **7. Plotting the elbow curve**

```python
#Now we make Elbow curve to find out the correct number of clusters. For that we consider the possibility from 1 to 10 clusters.
SSE = []
for cluster in range(1,10):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(iris1)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
```

```python
#Plotting the elbow curve.
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# This Elbow curve shows that the optimum number of clusters is '3'.
```

### **8. Fitting the species into the clusters**

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(iris1)
pred=kmeans.predict(iris1)
```

```python
iris1['cluster'] = pred
```

<div id='9'></div>

<br>

### **9. Visualization of clusters**

```python
def seg(str_x, str_y, clusters):
  x = []
  y = []

  for i in range(clusters):
    x.append(iris1[str_x][iris1['cluster']==i])
    y.append(iris1[str_y][iris1['cluster']==i])
  
  return x,y

def plot_clusters(str_x, str_y, clusters):
  plt.figure(figsize = (5,5), dpi = 120)
  
  x,y = seg(str_x, str_y, clusters)

  for i in range(clusters):
    plt.scatter(x[i], y[i], label = 'cluster{}'.format(i))
  
  plt.xlabel(str_x)
  plt.ylabel(str_y)
  plt.title(str(str_x+" Vs "+str_y))
  plt.legend()
```

```python
plot_clusters('SepalLengthCm', 'SepalWidthCm',3)
```

```python
plot_clusters('PetalLengthCm', 'PetalWidthCm',3)
```

### **10.  Centroid of the clusters**

```python
x = iris1.iloc[:, [0, 1, 2, 3]].values
# Visualising the clusters - On the first two columns
plt.scatter(x[pred == 0, 0], x[pred == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[pred == 1, 0], x[pred == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[pred == 2, 0], x[pred == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
